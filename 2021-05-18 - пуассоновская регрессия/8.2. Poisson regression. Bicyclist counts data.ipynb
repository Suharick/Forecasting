## 1. Чтение и подготовка данных

Рассмотрим данные о количестве велосипедистов. Количество велосипедистов зависит от погодных условий в рассматриваемый день: чем хуже погода, тем меньше желающих. В качестве признаков возьмем:
- максимальную температуру в рассматриваемый день (F);
- минимальную температуру в рассматриваемый день (F);
- количество осадков.

```python
data = pd.read_csv('data/nyc_bicyclist_counts.csv', index_col=['Date'], parse_dates=True)
```

```python
data.head()
```

Целевая переменная &ndash; `'BB_COUNT'` &ndash; содержит только целые положительные числа, что должно быть учтено при выборе предсказательной модели.

```python
data['BB_COUNT'].plot(figsize=(12,5))
plt.show()
```

Кроме указанных факторов, количество велосипедистов может зависеть от дня недели: в выходные количество желающих больше, нежели в будни. Также может оказаться важным месяц. Добавим столбцы, содержащие информацию о том, на какой день недели и на какой месяц приходится наблюдение:

```python
data['DAY_OF_WEEK'] = data.index.dayofweek
data['MONTH'] = data.index.month
```

Данные переменные являются категориальными.

```python
from scipy.optimize import minimize
from sklearn.metrics import mean_absolute_error
```

```python
data['one'] = ([1 for i in range(len(data))])
```

```python
data.head(10)
```

```python
X = data.drop(['BB_COUNT'], axis=1).values
Y = data['BB_COUNT'].values
```

```python
def MNK(X, Y, teta):
    res = np.dot(X, teta)
    return sum((res - Y)**2)
```

```python
def predict_MNK(X, teta):
    return np.dot(X, teta)
```

```python
res_MNK = minimize(lambda teta: MNK(X, Y, teta), (1, 1, 1, 1, 1, 1))
res_MNK_params = res_MNK.x
```

```python
y_pred_MNK = predict_MNK(X, res_MNK_params)
```

```python
mean_absolute_error(Y, y_pred_MNK)
```

```python
def LLF(X, Y, teta):
    res = np.dot(X, teta)
    return np.mean(np.exp(res) - Y*res)
```

```python
def predict_LLF(X, teta):
    return np.exp(np.dot(X, teta))
```

```python
res_LLF = minimize(lambda teta: LLF(X, Y, teta), (0, 0, 0, 0, 0, 0))
res_LLF_params = res_LLF.x
```

```python
y_pred_LLF = predict_LLF(X, res_LLF_params)
```

```python
mean_absolute_error(Y, y_pred_LLF)
```

```python
plt.plot(Y)
plt.plot(y_pred_MNK)
plt.plot(y_pred_LLF)
plt.show()
```

```python
X_new = pd.get_dummies(X, columns=['DAY_OF_WEEK', 'MONTH'])
```

```python
teta0 = np.ones(X_new.shape[1])
```

```python
lin_reg = minimize(lambda teta: MNK(X_new, Y, teta), tuple(teta0))
lin_reg_params = lin_reg.x
```

```python
teta0 = np.zeros(X_new.shape[1])
```

```python
llf_model = minimize(lambda teta: LLF(X_new, Y, teta), tuple(teta0))
llf_params = llf_model.x
```

```python
y_pred_lin = np.dot(X_new, lin_reg_params)
y_pred_llf = np.exp(np.dot(X_new, llf_params))
```

```python
mean_absolute_error(Y, y_pred_lin)
```

```python
mean_absolute_error(Y, y_pred_llf)
```
