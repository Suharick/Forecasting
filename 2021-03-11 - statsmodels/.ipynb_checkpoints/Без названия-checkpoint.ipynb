```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bus = pd.read_csv('data/weekday-bus-ridership.csv', index_col='Month', parse_dates=True)
bus
```

```python
bus.plot(legend=True)
plt.show()
```

Линия тренда появляется с середины 1977 года, до этого данные находятся примерно на горизонтальной линии тренда

Сезонность не идеальная, но присутствует: с середины каждого года идет прирост, далее недолгий спуск, после этого за короткий срок идет достижение максимального значения сезона, после чего резкий спад почти до той же отметки, с которой стартовал график.

<h3>Создадим прогноз на 2 года (измерение идет в неделях)</h3>

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
ses = ExponentialSmoothing(bus, trend = 'mul', seasonal = 'mul', seasonal_periods = 12, freq = 'MS').fit()
```

```python
print(f'Standard deviation\nInitial: {bus.values.std()}', f'\nFitted: {ses.fittedvalues.values.std()}')
```

```python
plt.figure(figsize=(12,4), dpi=150)
bus.plot(legend=True)
ses.fittedvalues.rename(r'$\alpha=%.2f$' % ses.model.params['smoothing_level'] + ', ' +
                        r'$l_0=%.4f$' % ses.model.params['initial_level']).plot(legend=True)
plt.show()
```

```python
ses_prognosis = ses.forecast(24)
```

```python
plt.figure(figsize=(12,4), dpi=150)
bus.plot(legend=True)
ses_prognosis.rename(r'$\alpha=%.2f$' % ses.model.params['smoothing_level'] + ', ' +
                     r'$l_0=%.4f$' % ses.model.params['initial_level']).plot(legend=True)
plt.show()
```
