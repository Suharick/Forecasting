## Определение порядка авторегрессии для своих данных

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
electricity = pd.read_csv('data/monthly-electricity-production.csv', index_col=0, parse_dates=True)
electricity.head(10)
```

```python
electricity.plot(figsize=(12,5))
plt.show()
```

### 1. Автокорреляция и частичная автокорреляция

Преобразуем ряд в стационарный, применив метод Бокса-Кокса, сезонное и обычное дифференцирование

```python
from scipy.stats import boxcox
from statsmodels.tsa.stattools import adfuller
```

```python
electricity_transformed, lmbda = boxcox(electricity['Monthly electricity production in Australia: million kilowatt hours. Jan 1956 ? Aug 1995'])
y_transformation = pd.DataFrame(electricity['Monthly electricity production in Australia: million kilowatt hours. Jan 1956 ? Aug 1995'].values, columns=['initial'])
y_transformation['transformed'] = electricity_transformed
y_transformation['seasonal_diff'] = y_transformation['transformed'].diff(12)
y_transformation['lag1_diff'] = y_transformation['seasonal_diff'].diff(1)
```

```python
print("Критерий Дики-Фуллера для исходного ряда: p=%f" % adfuller(y_transformation['initial'])[1])
print("После преобразования Бокса-Кокса: p=%f" % adfuller(y_transformation['transformed'])[1])
print("После сезонного дифференцирования: p=%f" % adfuller(y_transformation['seasonal_diff'].dropna())[1])
print("После дополнительного дифференцирования: p=%f" % adfuller(y_transformation['lag1_diff'].dropna())[1])
```

Рассмотрим коррелограмму полученного ряда.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plt.rcParams['figure.figsize'] = (12,5)
plot_acf(y_transformation['lag1_diff'].dropna(), lags=48)
plt.xticks(np.arange(0, 50, 2))
plt.show()
```

Рассмотрим коррелограмму частичной автокорреляции ряда.

```python
plt.rcParams['figure.figsize'] = (12,5)
plot_pacf(y_transformation['lag1_diff'].dropna(), lags=48)
plt.xticks(np.arange(0, 50, 2))
plt.show()
```

```python
p = [1, 2, 3]
q = [0, 1, 2]
P = [2, 3, 4]
Q = [0, 1, 2]
```

```python
from itertools import product
```

```python
parameters = product(p, q, P, Q)
parameters_list = list(parameters)
```

### 2. Построение модели SARIMA

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
minAIC1 = 10000000
opt_par1 = -1

for i in parameters_list:
    model = SARIMAX(y_transformation['initial'], 
                order=(i[0], 1, i[1]), 
                seasonal_order=(i[2], 1, i[3], 12)).fit()
    if minAIC1 > model.aic:
        minAIC1 = model.aic
        opt_par1 = i
    
opt_par1
```

```python
y_transformation['initial'].plot(figsize=(12,5))
model.forecast(24).plot()
plt.legend(['Initial', 'Forecast'])
plt.show()
```
