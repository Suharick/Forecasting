## Построение моделей прогнозирования на основе признаков

Для описания поведения ряда могут использоваться как экзогенные факторы, так и факторы, полученные непосредственно из временного ряда: моменты времени, индикаторы дней недели, статистики, посчитанные на данных из прошлого и т. д.

```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize

from matplotlib import pyplot as plt
plt.style.use('bmh')
plt.rcParams["figure.figsize"] = (15,8)
```

```python
data = pd.read_csv('passengers.csv', parse_dates=True)
data = data.set_index('date')
```

```python
data.head()
```

```python
data.plot()
```

## Задание 1

```python
# формируем матрицу признаков
val = []
n = data.shape[0]
for i in range(0, n):
    val.append([1, i, np.sin(2*np.pi*(i-1)/12), np.cos(2*np.pi*(i-1)/12)])
```

```python
def MSE(Q1, Q2, b1, b2):
    res = np.dot(val, np.array([[Q1],[Q2],[b1], [b2]]))
    t = len(val)
    return sum((res.reshape(t) - data["num_passengers"].values)**2)/t
```

```python
def get_prediction(Q1, Q2, b1, b2):
    return (np.dot(val, np.array([[Q1],[Q2],[b1], [b2]]))).reshape(len(val))
```

```python
# МНК: находим оптимальные параметры
result = minimize(lambda x: MSE(x[0], x[1], x[2], x[3]), (0.5, 0.5, 0.5, 0.5))
```

```python
Q1, Q2, b1, b2 = result.x
Q1, Q2, b1, b2
```

```python
y_pred = get_prediction(Q1,Q2,b1, b2)
```

```python
y_pred.shape
```

```python
plt.plot(y_pred)
plt.plot(data["num_passengers"].values)
plt.show()
```
